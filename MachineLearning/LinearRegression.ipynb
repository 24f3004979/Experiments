## Linear Regression using Least squares method
Making Linear Regression learning with least squares and simple algorithms using numpy

```python
import numpy as np
import matplotlib.pyplot as plt

SIZE = 10

x = np.linspace(-20,20,SIZE)
noice_scale = 1/8
noice = np.array(np.random.normal(noice_scale,1, x.size))

m, b  = 3.4, 2.8  # Equation parameters

y_target = m * x + (b * noice)
y = m * x + b

print(f" x is given as : {x}")

# Ploting Stuff
plt.scatter(x,y_target)
plt.scatter(x,y, color="red")
```

```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt


# generate x and y
x = np.linspace(0, 1, 101)
y = 1 + x + x * np.random.random(len(x))

# assemble matrix A
A = np.vstack([x, np.ones(len(x))]).T

# turn y into a column vector
y = y[:, np.newaxis]

# Direct least square regression
alpha = np.dot(
    (np.dot(np.linalg.inv(np.dot(A.T,A)), A.T))
    ,y)

y_predicted = alpha[0] * x  + alpha[1]
```

```python
# Visualising Plot
import matplotlib.pyplot as plt

plt.title = "Least Square"
plt.scatter(x,y, color = "blue")
plt.plot(x, y_predicted, color="red")
```
