```python
#@title Selection Sort

def SelectionSort(elements):
    '''
    Take minimum of the list and swap the first element '''
    if len(elements) <= 1:
        return elements
    
    minimum = elements[0]
    for element in elements:
        if minimum > element:
            minimum = element
    # Making position swap
    min_element_index = elements.index(minimum)
    elements[0],elements[min_element_index] = minimum, elements[0]
    return [elements[0]] + SelectionSort(elements[1:])
```

```python
#@title Insertion Sort

def InsertionSort(elements):
    '''
    Sorting just the last part of the list
    Not working due to none type concatination Error
    now Logic is broken :)
    '''
    if len(elements) == 1:
        return elements
    print(f"DEBUG Elements : {elements}")

    if elements[-1] < elements[-2]:
        elements[-1], elements[-2] = elements[-2],elements[-1]
        return [elements[-1]] + InsertionSort(elements[:-1])
    return [elements[-1]] + InsertionSort(elements[:-1])
```
