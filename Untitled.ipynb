```python
import networkx as nx
import matplotlib.pyplot as plt
import random

# Random Points
l = []
for i in range(1,10):
    x = random.randint(-200,200)
    y = random.randint(-20,200)
    l.append((x,y))


G = nx.Graph(l)


nx.draw(G, with_labels=True, node_color='lightblue', node_size=800)
plt.show()
```
