## Sorting Algorithms
Making simple sorting bare algorithms for Making elements sorting

1. Selection sorting
       ~ Making sorting of element by swaping the minimum of the given list in ordered way
       Iterating the list swaping the minimum of current list into the bigining position
2. Insertion Sorting
        Way to make the sorting with comparing two and making sorts
       Quite unclear about what it was really and merge sort is really with great complexity but lets see how it works out to manage the complexity

```python
#@title Selection Sort

def SelectionSort(elements):
    '''
    Take minimum of the list and swap the first element '''
    if len(elements) <= 1:
        return elements
    
    minimum = elements[0]
    for element in elements:
        if minimum > element:
            minimum = element
    # Making position swap
    min_element_index = elements.index(minimum)
    elements[0],elements[min_element_index] = minimum, elements[0]
    return [elements[0]] + SelectionSort(elements[1:])
```

```python
#@title Insertion Sort

def InsertionSort(elements):
    '''
    Sorting just the last part of the list
    Not working due to none type concatination Error
    now Logic is broken :)
    '''
    if len(elements) == 1:
        return elements
    print(f"DEBUG Elements : {elements}")

    if elements[-1] < elements[-2]:
        elements[-1], elements[-2] = elements[-2],elements[-1]
        return [elements[-1]] + InsertionSort(elements[:-1])
    return [elements[-1]] + InsertionSort(elements[:-1])
```
