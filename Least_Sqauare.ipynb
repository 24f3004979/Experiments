# Least Squares implementation

Making basic working linear regression and implementing least square into data set for making linear model out of datasets

```python
import numpy as np 
import matplotlib.pyplot as plt 

class LinearRegressionLS:
    def __init__(self):
        '''
        Making Linear Regression model
         + Core Attributes
             - Theta : Set of parameters to fit for :)
        + Basic Functions
            ~ Using Fit function for geting theta with Least Squares
            ~ Simple Plot for visualising the stuff
        '''
        
        self.theta = None

    def fit(self, X, y):
        '''
        Procedure
        + Make one stock for b parameter 

        X.T ( X transpose) @ X { Matrix Multiplication }
        Expression for minimizing error 
        
        '''
        X = np.column_stack((X, np.ones(len(X))))
        # Normal Equation
        self.theta = np.linalg.inv(X.T @ X) @ X.T @ y

    def predict(self, X):
        '''
        y = X @ theta
        Returning the prediction based on fit
        '''
        X = np.column_stack((X, np.ones(len(X))))
        return X @ self.theta
            
    def plot(self, X, y):
        plt.scatter(X, y)
        y_pred = self.predict(X)
        plt.plot(X, y_pred)
        plt.xlabel("Input sets X")
        plt.ylabel("Output sets y")
        plt.title("Linear Regression (Least sqaures)")
        plt.show()
        
```

```python
# Sample data
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

model = LinearRegressionLS()
model.fit(X, y)

print("w, b =", model.theta)

model.plot(X, y)
```

```python
# testing with big random data set
import numpy as np

def generate_linear_data(
    n_samples=100000,
    n_features=3,
    noise_std=1.0,
    random_seed=42
):
    """
    Generates synthetic linear regression data.

    y = Xw + b + noise

    Parameters:
        n_samples : number of data points
        n_features : number of input features
        noise_std : standard deviation of Gaussian noise
        random_seed : for reproducibility

    Returns:
        X : (n_samples, n_features)
        y : (n_samples,)
        true_theta : (n_features + 1,)
    """

    np.random.seed(random_seed)

    # Generate feature matrix
    X = np.random.randn(n_samples, n_features)

    # True weights and bias
    true_w = np.random.randn(n_features)
    true_b = np.random.randn()

    # Generate noise
    noise = np.random.randn(n_samples) * noise_std

    # Generate target
    y = X @ true_w + true_b + noise

    # Combine true parameters
    true_theta = np.append(true_w, true_b)

    return X, y, true_theta

x,y, true_theta = generate_linear_data()
model = LinearRegressionLS()
model.fit(X, y)

model.plot(X,y)
```
