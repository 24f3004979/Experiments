# BFS searching Procedure
Exploring whole vertice with using Queue based mechanism for exploring the graph

Exploring Graph with node, Adding element as visited and then exposing other elements of that element into queue for visiting, and moving respectivily, 
Expanding with the domain

Remove the visited part -> Add its All connected node
Proceed with other nodes to explore about

Time complexity : $$ n^2 $$
Best Time complexity of searching whole graph : $$ O(n + m)$$

Using BFS benifits
1. tracing path
2. tracking level
3. parent info -> path recorver
4. path computation with paths
