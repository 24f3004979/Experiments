# Quick Sorting algorithm
Procedure 
1. Take Pivot [ default first element]
2. Make partitions for lower and upper numbers
3. concatinate the final output

Analysis required for this algorithm


```python
# Implementation

def QuickSort(L):
    if len(L) <= 1:
        return L
    pivot = L[0]
    i = 1
    while i <= (len(L) - 1):
        element = L[i]
        compare = (element <= pivot)
        if compare:
            section = True
        else:
            pivot_position = i - 1 # previos as pivot position
            section = False
        if compare and (section == False):
            L[1], L[i] = element, L[1] # swaping first element with this minn
        i += 1
    # Seting pivot and next iteration 
    L[0], L[pivot_position] = L[pivot_position], L[0] # swap 
    print(f"Intermidiate List : {L}")
    return QuickSort(L[:pivot_position]) + pivot + QuickSort(L[pivot_position:])
            
```

```python
# Testing
L = [2,3,1,6]
print(f"Quick sorting : {QuickSort(L)}")
```
